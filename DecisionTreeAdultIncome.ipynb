```python
# -----------------------------------------------------------------
# Decision Tree Classifier
# Predict the income of an adult based on the census data
# -----------------------------------------------------------------

# Import libraries
import pandas as pd
```

```python
# Read dataset
data = pd.read_csv(r'C:\Users\o.resulov\Desktop\QSS\decisiontreeAdultIncome.csv')
data
```

```python
# Check for Null values
data.isnull().sum(axis=0)
```

```python
# Create Dummy variables
data.dtypes
data = pd.get_dummies(data, drop_first=True)
data
```

```python
# Create X and Y Variables
data.columns
y=data['IncomeClass_ >50K']
X=data.drop(['IncomeClass_ >50K'], axis=1)
X
```

```python
# Split the X and Y dataset into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, stratify=y)
```

```python
# Import and train classifier
from sklearn import tree
model = tree.DecisionTreeClassifier(criterion='entropy')
model.fit(X_train, y_train)
```

```python
# Test the model
model_estimation = model.predict(X_train)
model_estimation
```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_train, model_estimation)
f1 = accuracy_score(y_train, model_estimation)
```
